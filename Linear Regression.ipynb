# Linear Regression Using Scikit Learn

```python
# Import all dependencies
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Read the data
```

```python
data = pd.read_csv("D:\\DATA SCIENCE\\ALGORITHMS\\MACHINE LEARNING ALGORITHMS\\Salary_dataset SLM.csv")
```

```python
# Look at the data
```

```python
data.head()
```

```python
# Remove redundant columns
```

```python
data.drop(['Unnamed: 0'],axis=1)
```

```python
# Visualise the data (TV Ads and Sales)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    data['YearsExperience'],
    data['Salary'],
    c='red'
)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Generate Linear Approximation of this data
```

```python
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values.reshape(-1, 1)

reg = LinearRegression()
reg.fit(X, y)

print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
# Visualise the Best Fit Line
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(data['YearsExperience'],data['Salary'],c='red')
plt.plot(data['YearsExperience'],reg.predict(X),c='blue',linewidth=2)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
