```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
df=pd.read_csv("D:\\DATA SCIENCE\\ALGORITHMS\\MACHINE LEARNING ALGORITHMS\\salaries Decision tree.csv")
df
```

```python
df.shape
```

```python
inputs=df.drop("salary_more_then_100k",axis=1)
inputs
```

```python
target=df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
la_company=LabelEncoder()
la_job=LabelEncoder()
la_degree=LabelEncoder()
```

```python
inputs['company_n']=la_company.fit_transform(inputs['company'])
inputs['job_n']=la_job.fit_transform(inputs['job'])
inputs['degree_n']=la_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis=1)
```

```python
inputs_n
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[1,2,1]])
```

```python
plt.figure(figsize=(20,15))
tree.plot_tree(model,filled=True)
```
